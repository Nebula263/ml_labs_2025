2. Постройте гистограмму для аропортов по количеству отменённых рейсов. Выведите топ 10 самых злостных отменщиков. Столбики отсортировать по убыванию слева на право. Определите аэропорт, рейсы для которого отменяются наиболее часто.

3. Определите коэффициент корреляции Пирсона и Спирмена между отменой рейса и днём недели, месяцем. Оцените значение p-value. Постройте плотность распределения количества отмен по месяцам и по дням недели (используйте kdeplot или displot).

```python
from scipy.stats import pearsonr, spearmanr

pearson_day, p_day = pearsonr(df['DAY_OF_WEEK'], df['CANCELLED'])
spearman_day, _ = spearmanr(df['DAY_OF_WEEK'], df['CANCELLED'])
pearson_month, p_month = pearsonr(df['MONTH'], df['CANCELLED'])
spearman_month, _ = spearmanr(df['MONTH'], df['CANCELLED'])

print(f"Пирсон (день недели): {pearson_day:.3f}, p-value: {p_day:.3f}")
print(f"Спирмен (день недели): {spearman_day:.3f}")
print(f"Пирсон (месяц): {pearson_month:.3f}, p-value: {p_month:.3f}")
print(f"Спирмен (месяц): {spearman_month:.3f}")

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.kdeplot(data=df, x='MONTH', hue='CANCELLED', common_norm=False, legend = False)
plt.title('Распределение отмен по месяцам')

plt.subplot(1, 2, 2)
sns.kdeplot(data=df, x='DAY_OF_WEEK', hue='CANCELLED', common_norm=False, legend = False)
plt.title('Распределение отмен по дням недели')
plt.show()
```

4. Подсчитайте для трёх выбранных авиакомпаний: количество рейсов, количество отменённых рейсов, количество перенаправленных рейсов. По каждому типу постройте histplot.

5. Сформируйте новый признак скорости полёта для каждого рейса. Рассчитайте среднюю скорость полёта по всем рейсам для трёх выбранных авиакомпаний и выведите её на histplot.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df['SPEED'] = df['DISTANCE'] / (df['AIR_TIME'] / 60)

selected_airlines = ['UA', 'AA', 'US']

for airline in selected_airlines:
    airline_data = df[df['AIRLINE'] == airline]

    avg_speed = airline_data['SPEED'].mean()

    plt.figure(figsize=(10, 6))
    sns.histplot(airline_data['SPEED'], bins=20, kde=True)

    plt.title(f"Распределение скорости полёта для авиакомпании {airline}")
    plt.xlabel("Скорость (миль/ч)")
    plt.ylabel("Частота")
    plt.show()

    print(f"Средняя скорость полёта для авиакомпании {airline}: {avg_speed:.2f} миль/ч")
```

6. Возьмите 10 дней из общей выборки. Визуализируйте тепловую карту (heatplot) частоты отмены рейсов. По одной оси – дни, по другой оси – рейс (для двух аэропортов).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df['DATE'] = pd.to_datetime(df[['YEAR', 'MONTH', 'DAY']])


days = df['DATE'].drop_duplicates().sample(10, random_state=1)
airports = df['ORIGIN_AIRPORT'].value_counts().head(2).index


small_df = df[
    df['DATE'].isin(days) &
    df['ORIGIN_AIRPORT'].isin(airports)
]


heat_data = small_df.groupby(['DATE', 'ORIGIN_AIRPORT'])['CANCELLED'].mean().unstack()


plt.figure(figsize=(12, 6))
sns.heatmap(heat_data.T,
           cmap='YlOrRd',
           annot=True,
           fmt='.1%',
           linewidths=0.5,
           cbar_kws={'label': 'Доля отмен'})

plt.title('Тепловая карта отмен рейсов по дням и аэропортам', pad=20)
plt.xlabel('Дата')
plt.ylabel('Аэропорт отправления')
plt.xticks(rotation=45)

date_labels = heat_data.index.strftime('%Y-%m-%d')
plt.xticks(ticks=range(len(date_labels)), labels=date_labels)

plt.tight_layout()
plt.show()
```

7. Сформируйте новый датасет по авиакомпаниям с признаками: среднее время задержки отправки и среднее время прибытия.

```python
delay_stats = df.groupby('AIRLINE').agg({
    'DEPARTURE_DELAY': 'mean',
    'ARRIVAL_DELAY': 'mean'
}).reset_index()

delay_stats.columns = ['AIRLINE', 'MEAN_DEP_DELAY', 'MEAN_ARR_DELAY']
display(delay_stats.head())
```

8. Постройте два histplot по задержкам отправления и прибытия для топа 5 компаний.

```python
import seaborn as sns
import matplotlib.pyplot as plt


top5 = df['AIRLINE'].value_counts().nlargest(5).index
top5_df = df[df['AIRLINE'].isin(top5)]

sns.histplot(data=top5_df, x='DEPARTURE_DELAY', hue='AIRLINE', element='step', stat='density', common_norm=False)
plt.title("Задержка отправки – топ 5")
plt.xlim(-20, 200)
plt.show()

sns.histplot(data=top5_df, x='ARRIVAL_DELAY', hue='AIRLINE', element='step', stat='density', common_norm=False)
plt.title("Задержка прибытия – топ 5")
plt.xlim(-20, 200)
plt.show()
```
